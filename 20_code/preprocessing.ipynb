```python
national = pd.read_csv("../10_data/national20212022_PUBLIC_state.csv")
national
```

```python
national_subset = national[columns_kept]
```

```python
# Gender variables
national_subset.loc[national_subset['gender_male'].isna(), 'gender_male'] = 0
national_subset.loc[national_subset['gender_female'].isna(), 'gender_female'] = 0
national_subset.loc[national_subset['gender_transm'].isna(), 'gender_transm'] = 0
national_subset.loc[national_subset['gender_transf'].isna(), 'gender_transf'] = 0
national_subset.loc[national_subset['gender_nonbin'].isna(), 'gender_nonbin'] = 0
national_subset.loc[national_subset['gender_selfid'].isna(), 'gender_selfid'] = 0

# Sexuality variables
national_subset.loc[national_subset['sexual_h'].isna(), 'sexual_h'] = 0
national_subset.loc[national_subset['sexual_l'].isna(), 'sexual_l'] = 0
national_subset.loc[national_subset['sexual_g'].isna(), 'sexual_g'] = 0
national_subset.loc[national_subset['sexual_bi'].isna(), 'sexual_bi'] = 0
national_subset.loc[national_subset['sexual_queer'].isna(), 'sexual_queer'] = 0
national_subset.loc[national_subset['sexual_quest'].isna(), 'sexual_quest'] = 0
national_subset.loc[national_subset['sexual_asexual'].isna(), 'sexual_asexual'] = 0
national_subset.loc[national_subset['sexual_pan'].isna(), 'sexual_pan'] = 0
national_subset.loc[national_subset['sexual_other'].isna(), 'sexual_other'] = 0


# Race variables
national_subset.loc[national_subset['race_black'].isna(), 'race_black'] = 0
national_subset.loc[national_subset['race_ainaan'].isna(), 'race_ainaan'] = 0
national_subset.loc[national_subset['race_asian'].isna(), 'race_asian'] = 0
national_subset.loc[national_subset['race_his'].isna(), 'race_his'] = 0
national_subset.loc[national_subset['race_pi'].isna(), 'race_pi'] = 0
national_subset.loc[national_subset['race_mides'].isna(), 'race_mides'] = 0
national_subset.loc[national_subset['race_white'].isna(), 'race_white'] = 0
national_subset.loc[national_subset['race_other'].isna(), 'race_other'] = 0
```

```python
# First-gen variables
national_subset['first_gen'] = (national_subset['educ_par1'] < 4) & (national_subset['educ_par2'] < 4)
national_subset.loc[(national_subset['educ_par1'].isna()) & (national_subset['educ_par2'].isna()), 'first_gen'] = np.nan
```

```python
national_subset.loc[national_subset['scholarship_status_1'] == 1, 'scholarship_status_1'] = 0
national_subset.loc[national_subset['scholarship_status_1'].isna(), 'scholarship_status_1'] = 1
```
