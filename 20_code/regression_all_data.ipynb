### Start with regression with only `C(in_online)` and both `C(in_online) + cases_per_10k`
> Without any control variables, the results now seems reasonable and statistically significant.

> After adding cases_per_10k, the coefficients for `is_online` is still reasonable, but i feel like the coefficient for cases_per_10k should be positive ---> adding more control variables

```python
# wls regression
smf.wls('anxiety ~ C(is_online)', data=all_data).fit().summary()
```

```python
smf.wls('anxiety ~ C(is_online)+cases_per_10k', data=all_data).fit().summary()
```

### `C(in_online)` + `cases_per_10k` AND `enroll`

```python
smf.wls('anxiety ~ C(is_online)+cases_per_10k+ C(enroll)', data=all_data).fit().summary()
```

```python
# add interaction term
smf.wls('anxiety ~ C(is_online)*C(enroll)+cases_per_10k', data=all_data).fit().summary()
```

> `C(is_online)[T.1.0]` is the difference between full_time + on_line/inperson --- positive+significant ✅

> `C(is_online)[T.1.0]:C(enroll)[T.2.0]+C(is_online)[T.1.0]`is the difference between part_time + on_line/inperson(compared to baseline full_time) ---negative(less anxiety)+significant ✅
